### 1.3 3. For each of the categorical variables, construct a bar chart of the variable, with an overlay of the target variable. Normalize if necessary.

```python
# crosstab to find the count of <=50K and >50K for 'workclass'

crossTab = pd.crosstab(dataset['workclass'], dataset['income'])

# normalize the data
crossTab_norm = crossTab.div(crossTab.sum(axis=1), axis=0)

# plot the data -- without normalization
fig, ax = plt.subplots(figsize=(12, 6))
crossTab.plot(kind='bar', stacked=True, ax=ax, color=['blue', 'green']) 
ax.set_xlabel('Workclass')
ax.set_ylabel('Frequency')
ax.set_title('Non-normalized Bar Chart of Workclass')
ax.legend(title= 'income', loc='upper right')

# plot the data -- with normalization
fig, ax = plt.subplots(figsize=(12, 6))
crossTab_norm.plot(kind='bar', stacked=True, ax=ax, color=['blue', 'green'])

ax.set_xlabel('Workclass')
ax.set_ylabel('Proportion')
ax.set_title('Normalized Bar Chart of Workclass')
ax.legend(title= 'income', loc='upper right')

plt.tight_layout()
plt.show()
```
